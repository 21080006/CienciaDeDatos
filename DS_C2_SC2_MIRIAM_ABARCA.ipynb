f. Determinar la cantidad de vacunas aplicadas el d칤a 29/01/21 en todo el mundo.

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicaci칩n entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
conDiferencias = vaccinations
conDiferencias["diferencias"] = (conDiferencias["daily_vaccinations"] - conDiferencias["daily_vaccinations_raw"]).abs()
conDiferencias
```

h. Obtener el periodo de tiempo entre el registro con fecha m치s reciente y el registro con fecha m치s antigua.

```python
print("Fecha de registro mas reciente" ,vaccinations["date"].max())
print("Fecha de registro con fecha mas antigua", vaccinations["date"].min())
print("Perido de tiempo entre la fecha mas antigua hasta la mas reciente", vaccinations["date"].max() - vaccinations["date"].min())
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada d칤a (usar la columna vaccines y separar por el car치cter , ).

```python
conCantidad = vaccinations
conCantidad["canVac"] = conCantidad["vaccines"].str.split(",").str.len()
conCantidad
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vaccinations[vaccinations['date'] < '2020-12-20']
antes20
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vaccinations[vaccinations['vaccines'].str.contains("Pfizer")]
pfizer
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentaci칩n de pd.ExcelWriter.

```python
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
  conCantidad.to_excel(writer, sheet_name='conDiferencias')
  conDiferencias.to_excel(writer, sheet_name='conCantidad')
  antes20.to_excel(writer, sheet_name='antes20')
  pfizer.to_excel(writer, sheet_name='pfizer')
```
